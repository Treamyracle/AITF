```python
import pandas as pd
import numpy as np
import re

url = 'https://raw.githubusercontent.com/Treamyracle/AITF/refs/heads/main/%233/cleaning_dataset.csv'
df = pd.read_csv(url)

df
```

```python
def clean_column_names(df_input):
    cleaned_columns = []
    for col in df_input.columns:
        col = col.lower()
        col = re.sub(r'[^a-z0-9]+', '_', col)
        col = col.strip('_')
        cleaned_columns.append(col)

    df_input.columns = cleaned_columns
    return df_input

df = clean_column_names(df)
df_cleaned = df
df
```

```python
def clean_currency(value):
    if isinstance(value, str):
        value = re.sub(r'[\$,]', '', value)
        value = re.sub(r'\[.*?\]', '', value)
        return pd.to_numeric(value, errors='coerce')
    return value

currency_cols = ['actual_gross', 'adjusted_gross_in_2022_dollars', 'average_gross']
for col in currency_cols:
    df_cleaned[col] = df_cleaned[col].apply(clean_currency)

def clean_peak_column(value):
    if isinstance(value, str):
        match = re.match(r'(\d+)', value)
        if match:
            return int(match.group(1))
    return pd.to_numeric(value, errors='coerce')

df_cleaned['all_time_peak'] = df_cleaned['all_time_peak'].apply(clean_peak_column)

df_cleaned['tour_title'] = df_cleaned['tour_title'].str.replace(r'[†‡*]', '', regex=True).str.strip()

df_cleaned
```

```python
def clean_ranking_column(value):
    if isinstance(value, str):
        match = re.match(r'(\d+)', value)
        if match:
            return int(match.group(1))
    return pd.to_numeric(value, errors='coerce')

for col in ['peak', 'all_time_peak']:
    df_cleaned[col] = df_cleaned[col].apply(clean_ranking_column)

df_cleaned
```

```python
df_cleaned.drop(columns=['ref'], inplace=True)

for col in ['peak', 'all_time_peak']:
    median_val = df_cleaned[col].median()
    df_cleaned[col].fillna(median_val, inplace=True)

for col in currency_cols:
    mean_val = df_cleaned[col].mean()
    df_cleaned[col].fillna(mean_val, inplace=True)

df_cleaned
```
