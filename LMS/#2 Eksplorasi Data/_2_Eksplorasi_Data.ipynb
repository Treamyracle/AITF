```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df2 = pd.read_csv("https://raw.githubusercontent.com/Treamyracle/AITF/refs/heads/main/%231%20Data%20Scraping/student_exam_scores.csv")

print("\n1. Sampel Data (5 Baris Pertama):")
print(df2.head())

print("\n2. Dimensi Data (Baris, Kolom):")
print(df2.shape)

print("\n3. Tipe Data dan Informasi Umum:")
df2.info()

print("\n4. Statistik Deskriptif (Kolom Numerik):")
print(df2.describe())

print("\n5. Statistik Deskriptif (Kolom Kategorikal):")
print(df2.describe(include=['object']))

print("\n6. Visualisasi Distribusi Skor Ujian:")
plt.style.use('seaborn-v0_8-whitegrid')
plt.figure(figsize=(10, 5))
sns.histplot(df2['exam_score'], kde=True, bins=20)
plt.title('Distribusi Skor Ujian (Grade Class)')
plt.xlabel('Skor Ujian')
plt.ylabel('Frekuensi')
plt.show()
```

```python
print("\n1. Jumlah Nilai Hilang (Missing Values) per Kolom:")
missing_values = df2.isnull().sum()
if missing_values.sum() > 0:
    print(missing_values[missing_values > 0])
else:
    print("Tidak ditemukan nilai hilang dalam dataset.")

print("\n2. Jumlah Baris Duplikat:")
duplicate_rows = df2.duplicated().sum()
print(f"Ditemukan {duplicate_rows} baris duplikat.")

print("\n3. Validasi Nilai Unik pada Kolom Kategorikal:")
categorical_cols = df2.select_dtypes(include=['object']).columns
for col in categorical_cols:
    unique_values = df2[col].unique()
    print(f"  - Kolom '{col}' memiliki {len(unique_values)} nilai unik. Contoh: {unique_values[:5]}")


print(f"  - Rentang Skor Ujian: {df2['exam_score'].min():.2f} - {df2['exam_score'].max():.2f}")
if df2['exam_score'].min() < 0.0 or df2['exam_score'].max() > 100.0:
    print("    -> Peringatan: Ditemukan rentang Skor Ujian yang tidak wajar!")
```
